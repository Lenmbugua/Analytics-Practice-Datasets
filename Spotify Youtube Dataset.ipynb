```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Spotify Youtube Dataset.csv")
```

```python
df.head()
```

```python
df.columns       # started by checking the columns that we have
```

```python
df.info()     # Get some basic information of the dataset
```

```python
df.drop(columns = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace = True)    # drop the columns that are not important in analysing the dataset
```

```python
df.head()      # check the number of columns after droping the columns we do not need
```

```python
df.isna().sum()        # check the sum of missing values in each column
```

```python
# Fill the Likes and comments columns with 0
df['Likes'] = df['Likes'].fillna(0)
df['Comments'] = df['Comments'].fillna(0)
```

```python
df.isna().sum()    # run this to check if the columns have been filled.
```

```python
df.dropna(inplace = True)    # Drop the missing values of all the other columns
```

```python
df.isna().sum()     # check if the columns have been dropped
```

**Q.4) How do the Average Views, Likes, and Comments are compared between different Album Types?**

```python
data = df.groupby('Album_type')[['Likes', 'Views', 'Comments']].mean()
data
```

```python
type(data)
```

```python
data.reset_index()    #reset_index is used to convert the index of a Series into a column to form a DataFrame
```

```python
data = data.reset_index()
print(data.columns.tolist())
```

```python
#melt function used to unpivot a DataFrame
df_melted = pd.melt(data,id_vars='Album_type',var_name='Attribute',value_name='Total')
```

```python
df_melted
```

```python
#Draw the bar plot
plt.figure(figsize = (12,8))
sns.barplot(x = 'Album_type', y = 'Total', hue = 'Attribute', data = df_melted);
```

**Q.5) Top 5 YouTube Channels - based on the Views?**

```python
df.head(1)
```

```python
c_views = df.groupby('Channel')['Views'].sum().sort_values(ascending = False).head()
c_views
```

```python
c_views = c_views.reset_index()
c_views.head(10)
```

```python
# Draw a bar plot for this question
#sns.set_style("White")
sns.barplot(x = "Views", y = "Channel", data= c_views, color = 'Magenta');
plt.title('Top 5 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```

**Q.9) What is the Correlation between Views, Likes, Comments and streams**

```python
df.head(1)
```

```python
data_vlcs = df[['Views','Likes','Comments','Stream']]
data_vlcs
```

```python
data_vlcs.corr()      # used to show the corelation matrix
```

```python
sns.heatmap(data_vlcs.corr());     # using a heatmap to show the corelation matrix between Views, Likes, Comments and streams
# Corelation lies between 1 and -1. 1 means the two columns are highly corelated with each other while -1 means the two are not highly corelated.
# in the matrix below likes and views are highly corelated meaning if there are many views meaning the likes will also be more.
# stream and comments are not highly corelated meaning the less the streams the less the comments will be.
```
